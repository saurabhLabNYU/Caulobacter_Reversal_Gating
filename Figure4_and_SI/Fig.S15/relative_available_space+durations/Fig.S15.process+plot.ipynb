```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import expon
import re
```

```python
# excel
data_path = './WT_unique_pairs_relabeled.xlsx'
data = pd.read_excel(data_path)
# remove if 'Abonormal' or 'Blocked' or 'Brownian' or 'short' or 'test' in relabeling_notes
data = data[~data['relabeling_notes'].str.contains('Abnormal|Blocked|Brownian|short|test', na=False)]
data.columns
```

```python
# split based on Capsule_size
print(data['Capsule_size'].unique())
data_large = data[data['Capsule_size'] == 'Large']
data_small = data[data['Capsule_size'] == 'Small']
data_middle = data[data['Capsule_size'] == 'Middle']
print('number of large:', len(data_large))
print('number of small:', len(data_small))
print('number of middle:', len(data_middle))
```

```python
# get inner_diameter/body_length
dimension_ratio_large = data_large['inner_diameter/body_length']
dimension_ratio_small = data_small['inner_diameter/body_length']
# convert to numpy array
dimension_ratio_large = np.array(dimension_ratio_large)
dimension_ratio_small = np.array(dimension_ratio_small)
# convert to pd dataframe
dimension_ratio_large = pd.DataFrame(dimension_ratio_large, columns=['dimension_ratio'])
dimension_ratio_small = pd.DataFrame(dimension_ratio_small, columns=['dimension_ratio'])
```

```python
color_large = '#FF7F0E'  # orange
color_small = '#1F77B4'  # blue
# color_large = '#1F77B4'  # blue
# color_small = '#FF7F0E'  # orange

# not combine, calculate Fraction backward for each pair
data_current = dimension_ratio_large.copy()
# prepare data
# remove outliers (abs diff > 3 std)
data_current = data_current[(np.abs(data_current['dimension_ratio'] - data_current['dimension_ratio'].mean()) <= (3 * data_current['dimension_ratio'].std()))]
# plot dimension_ratio as hist
plt.figure(figsize=(5, 5))
plt.hist(data_current['dimension_ratio'].dropna(), bins=7, color=color_large, alpha=0.4, density=True, label='Large')
from scipy import stats
# add Gaussian fit
mu, std = stats.norm.fit(data_current['dimension_ratio'].dropna())
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p, color_large, linewidth=4)
print('N=', len(data_current['dimension_ratio'].dropna()))
print('mean=', data_current['dimension_ratio'].mean().round(2), 'std=', data_current['dimension_ratio'].std().round(2), 'median=', data_current['dimension_ratio'].median().round(2))
# add text to show mean and std
plt.text(5, 1.01, f'%.1f ± %.1f $L$' % (data_current['dimension_ratio'].mean(), data_current['dimension_ratio'].std()), color='k', fontsize=14)


data_current = dimension_ratio_small.copy()
# prepare data
# remove outliers (abs diff > 3 std)
data_current = data_current[(np.abs(data_current['dimension_ratio'] - data_current['dimension_ratio'].mean()) <= (3 * data_current['dimension_ratio'].std()))]
plt.hist(data_current['dimension_ratio'].dropna(), bins=7, color=color_small, alpha=0.4, density=True, label='Small')
# add Gaussian fit
mu, std = stats.norm.fit(data_current['dimension_ratio'].dropna())
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p, color_small, linewidth=4)
print('N=', len(data_current['dimension_ratio'].dropna()))
print('mean=', data_current['dimension_ratio'].mean().round(2), 'std=', data_current['dimension_ratio'].std().round(2), 'median=', data_current['dimension_ratio'].median().round(2))
# add text to show mean and std
plt.text(5, 0.932, f'%.1f ± %.1f $L$' % (data_current['dimension_ratio'].mean(), data_current['dimension_ratio'].std()), color='k', fontsize=14)

# add legend
plt.legend(loc='upper left', fontsize=12, frameon=False)



# label
plt.xlabel('Available Space $D_{in}$\n/Swimmer Length $L$', fontsize=16)
plt.ylabel('Probability Density', fontsize=16)

# x lim
plt.xlim(1.7, 6.4)
plt.ylim(0, 1.1)

# # --- dashed line of other bias---
# # open field bias: 0.25
# # outside bias: 0.33
# plt.axvline(x=0.25, color='black', linestyle='--', linewidth=1.5)
# plt.axvline(x=0.33, color='black', linestyle='--', linewidth=1.5)

# plt.figure(figsize=(6,5.5))

# ---Parameters---
ax = plt.gca()
# Font sizes
ax.tick_params(axis='both', which='major', labelsize=16)  # Tick label size

# Get the current content of the labels and title
xlabel = ax.get_xlabel()
ylabel = ax.get_ylabel()
title = ax.get_title()

# Set the font sizes without changing the content
ax.set_xlabel(xlabel, fontsize=16)  # X-axis label size
ax.set_ylabel(ylabel, fontsize=16)  # Y-axis label size
ax.set_title(title, fontsize=18)  # Title size

# Set legend size (if you have a legend)
legend = ax.get_legend()
if legend:
    for label in legend.get_texts():
        label.set_fontsize(14)

# Set text size for any additional annotations
for text in ax.texts:
    text.set_fontsize(14)

# Make axes thicker
for spine in ['top', 'right', 'bottom', 'left']:
    ax.spines[spine].set_linewidth(1.5)

# Reduce num of ticks
# ax.locator_params(axis='y', nbins=6)
# ax.locator_params(axis='x', nbins=4)
# set x ticks
plt.xticks([2, 4, 6], ['2', '4', '6'], fontsize=14)
# set y ticks
plt.yticks([0, 0.4, 0.8], ['0', '0.4', '0.8'], fontsize=14)

## Save figure as PNG (optional), transparent background, tight bounding box
# plt.savefig('large_vs_small_dimension_ratio.png', format='png', dpi=900, transparent=True, bbox_inches='tight')
```

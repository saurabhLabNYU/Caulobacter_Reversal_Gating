```python
# loop for TIRF488
# create a table to store the information
df_TIRF488 = pd.DataFrame(columns=['intensity', 'time', 'track_min_speed', 'track_max_speed', 'track_mean_speed', 'track_std_speed', 'track_median_speed', 'total_distance', 'mean_straight_line_speed', 'mean_directional_change_rate'])
for i in range(len(tracks_list_TIRF488)):
    file_path = tracks_list_TIRF488[i]
    # get the name of the file
    file_name = os.path.basename(file_path)
    # get intensity info and time info
    name_splits = file_name.split('_')
    intensity = name_splits[1]
    time = name_splits[3]
    print('intensity:', intensity)
    print('time:', time)
    # load 2D trajectory
    bug_2D_path = file_path
    # skip first 4 rows
    bug_2D = pd.read_csv(bug_2D_path, header=None, skiprows=4)
    # header is the original second row
    bug_2D.columns =  pd.read_csv(bug_2D_path, header=None).iloc[1]
    # get track min, max, mean, std, median speed
    track_min_speed = bug_2D['Track min speed'][0]
    track_max_speed = bug_2D['Track max speed'][0]
    track_mean_speed = bug_2D['Track mean speed'][0]
    track_std_speed = bug_2D['Track std speed'][0]
    track_median_speed = bug_2D['Track median speed'][0]
    # get Total distance traveled, Mean straight line speed, Mean directional change rate
    total_distance = bug_2D['Total distance traveled'][0]
    mean_straight_line_speed = bug_2D['Mean straight line speed'][0]
    mean_directional_change_rate = bug_2D['Mean directional change rate'][0]
    # append to the table
    df_current = pd.DataFrame([[intensity, time, track_min_speed, track_max_speed, track_mean_speed, track_std_speed, track_median_speed, total_distance, mean_straight_line_speed, mean_directional_change_rate]], columns=['intensity', 'time', 'track_min_speed', 'track_max_speed', 'track_mean_speed', 'track_std_speed', 'track_median_speed', 'total_distance', 'mean_straight_line_speed', 'mean_directional_change_rate'])
    df_TIRF488 = pd.concat([df_TIRF488, df_current], ignore_index=True)

# create a hyper table to store for each intensity
df_TIRF488_hyper = {}
intensity_list = df_TIRF488['intensity'].unique()
for intensity in intensity_list:
    # get the rows with intensity
    df_hyper_current = df_TIRF488[df_TIRF488['intensity'] == intensity]
    # add in the hyper table
    df_TIRF488_hyper[intensity] = df_hyper_current
```
